```python
#NUMBER OF COLLISIONS HAPPENING BY TIME OF DAY
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('https://data.cityofnewyork.us/resource/h9gi-nx95.csv')  
# Convert 'crash_time' to datetime format
data['crash_time'] = pd.to_datetime(data['crash_time'])

# Extract hour from 'crash_time'
data['hour_of_day'] = data['crash_time'].dt.hour

# Count occurrences of each hour of the day
hour_counts = data['hour_of_day'].value_counts()

# Sort by hour
hour_counts = hour_counts.sort_index()

# Plot the distribution of collisions by hour of the day
plt.figure(figsize=(10, 6))
hour_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Collisions by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Collisions')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
url = "https://data.cityofnewyork.us/resource/h9gi-nx95.csv"
data = pd.read_csv(url)

# Filter data for the specified boroughs
boroughs_to_compare = ['BROOKLYN', 'BRONX', 'MANHATTAN', 'QUEENS', 'STATEN ISLAND']
data_filtered = data[data['borough'].isin(boroughs_to_compare)]

# Visualizations for each borough
for borough in boroughs_to_compare:
    borough_data = data_filtered[data_filtered['borough'] == borough]
    
    # Bar plot of contributing factors for vehicle 1 by borough
    plt.figure(figsize=(12, 8))
    sns.countplot(x='contributing_factor_vehicle_1', data=borough_data, palette='muted')
    plt.title(f'Contributing Factors for Vehicle 1 in {borough}')
    plt.xlabel('Contributing Factors')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
    
    # Bar plot of contributing factors for vehicle 2 by borough
    plt.figure(figsize=(12, 8))
    sns.countplot(x='contributing_factor_vehicle_2', data=borough_data, palette='muted')
    plt.title(f'Contributing Factors for Vehicle 2 in {borough}')
    plt.xlabel('Contributing Factors')
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.show()
    
    # Check if there are fatalities data for this borough
    if 'number_of_persons_killed' in borough_data.columns:
        fatalities_data = borough_data[borough_data['number_of_persons_killed'] > 0]
        if not fatalities_data.empty:
            # Bar plot of fatalities by borough
            plt.figure(figsize=(10, 6))
            sns.countplot(x='number_of_persons_killed', data=fatalities_data, palette='viridis')
            plt.title(f'Fatalities in Auto Accidents in {borough}')
            plt.xlabel('Number of Fatalities')
            plt.ylabel('Count')
            plt.xticks(rotation=45)
            plt.show()
        else:
            print(f"No fatalities data available for {borough}")
    else:
        print(f"No fatalities data available for {borough}")
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "https://data.cityofnewyork.us/resource/h9gi-nx95.csv"  
# Adjust the file path as needed
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()

# Data Description
from IPython.display import display, HTML

display(HTML('<h1 style="color:blue; font-family:Verdana;">1. Data Description</h1>'))
display(HTML('<p style="font-family:Arial;">The dataset contains information about motor vehicle collisions in New York City. The key columns in the dataset include:</p>'))
display(HTML('''
<ul style="font-family:Arial;">
<li><strong>crash_date</strong>: Date of the crash</li>
<li><strong>crash_time</strong>: Time of the crash</li>
<li><strong>borough</strong>: Borough where the crash occurred</li>
<li><strong>zip_code</strong>: ZIP code of the crash location</li>
<li><strong>latitude</strong> and <strong>longitude</strong>: Geographic coordinates of the crash</li>
<li><strong>location</strong>: Combined latitude and longitude</li>
<li><strong>on_street_name</strong>, <strong>off_street_name</strong>, <strong>cross_street_name</strong>: Streets involved in the crash</li>
<li><strong>number_of_persons_injured</strong>, <strong>number_of_persons_killed</strong>: Number of persons injured or killed</li>
<li><strong>number_of_pedestrians_injured</strong>, <strong>number_of_pedestrians_killed</strong>: Number of pedestrians injured or killed</li>
<li><strong>number_of_cyclist_injured</strong>, <strong>number_of_cyclist_killed</strong>: Number of cyclists injured or killed</li>
<li><strong>number_of_motorist_injured</strong>, <strong>number_of_motorist_killed</strong>: Number of motorists injured or killed</li>
<li><strong>contributing_factor_vehicle_1</strong>, <strong>contributing_factor_vehicle_2</strong>: Contributing factors of the vehicles involved</li>
<li><strong>vehicle_type_code1</strong>, <strong>vehicle_type_code2</strong>: Types of vehicles involved</li>
</ul>
'''))

# Data Cleaning
display(HTML('<h1 style="color:green; font-family:Verdana;">2. Data Cleaning</h1>'))
display(HTML('<p style="font-family:Arial;">Steps for cleaning the data:</p>'))
display(HTML('''
<ol style="font-family:Arial;">
<li><strong>Handling Missing Values</strong>: Check for missing values in key columns. Fill missing borough values based on zip code if possible.</li>
<li><strong>Data Type Conversion</strong>: Convert date and time columns to appropriate data types.</li>
<li><strong>Removing Duplicates</strong>: Check and remove any duplicate rows.</li>
</ol>
'''))

# Handling Missing Values
missing_values = data.isnull().sum()
display(HTML('<h2 style="color:green; font-family:Verdana;">2.1 Handling Missing Values</h2>'))
print("Missing values in each column:\n", missing_values)

# Convert date and time columns to datetime types
data['crash_date'] = pd.to_datetime(data['crash_date'])
data['crash_time'] = pd.to_datetime(data['crash_time'], format='%H:%M').dt.time

# Fill missing borough values based on zip code (if possible)
# Assuming we have a mapping of zip codes to boroughs (zip_borough_map)
# data['borough'].fillna(data['zip_code'].map(zip_borough_map), inplace=True)

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Data Comparison by Borough
display(HTML('<h1 style="color:purple; font-family:Verdana;">3. Data Comparison by Borough</h1>'))

# Number of Accidents by Borough
accidents_by_borough = data['borough'].value_counts()
display(HTML('<h2 style="color:purple; font-family:Verdana;">3.1 Number of Accidents by Borough</h2>'))
print("Number of accidents by borough:\n", accidents_by_borough)

# Number of Injuries and Fatalities by Borough
injuries_fatalities_by_borough = data.groupby('borough')[['number_of_persons_injured', 'number_of_persons_killed',
                                                          'number_of_pedestrians_injured', 'number_of_pedestrians_killed',
                                                          'number_of_cyclist_injured', 'number_of_cyclist_killed',
                                                          'number_of_motorist_injured', 'number_of_motorist_killed']].sum()

display(HTML('<h2 style="color:purple; font-family:Verdana;">3.2 Number of Injuries and Fatalities by Borough</h2>'))
print("Injuries and fatalities by borough:\n", injuries_fatalities_by_borough)

# Visualization of Accidents
display(HTML('<h1 style="color:orange; font-family:Verdana;">4. Visualization of Accidents</h1>'))

# Bar plot for number of accidents by borough
plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_by_borough.index, y=accidents_by_borough.values, palette="viridis")
plt.title('Number of Accidents by Borough')
plt.xlabel('Borough')
plt.ylabel('Number of Accidents')
plt.show()

# Heatmap for injuries and fatalities by borough
plt.figure(figsize=(12, 8))
sns.heatmap(injuries_fatalities_by_borough, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Injuries and Fatalities by Borough')
plt.xlabel('Type')
plt.ylabel('Borough')
plt.show()
```
